```python
import pandas as pd # importing pandas libary
import numpy as np #importing numpy libary
import seaborn as sns #importing seaborn libary
import matplotlib.pyplot as plt #importing matplotlib libary
dfb1=pd.read_csv("horseasses-population-in-nepal-by-district.csv")
dfb2=pd.read_csv("milk-animals-and-milk-production-in-nepal-by-district.csv")
dfb3=pd.read_csv("net-meat-production-in-nepal-by-district.csv")
dfb4=pd.read_csv("production-of-cotton-in-nepal-by-district.csv")
dfb5=pd.read_csv("production-of-egg-in-nepal-by-district.csv")
dfb6=pd.read_csv("rabbit-population-in-nepal-by-district.csv")
dfb7=pd.read_csv("wool-production-in-nepal-by-district.csv")
dfb8=pd.read_csv("yak-nak-chauri-population-in-nepal-by-district.csv")
```

```python
# we capitalize the first letter using capitalize() function and remove unwanted space using strip()
dfb1['DISTRICT'] = dfb1['DISTRICT'].apply(lambda x : x.strip().capitalize())
dfb2['DISTRICT'] = dfb2['DISTRICT'].apply(lambda x : x.strip().capitalize())
dfb3['DISTRICT'] = dfb3['DISTRICT'].apply(lambda x : x.strip().capitalize())
dfb4['DISTRICT'] = dfb4['DISTRICT'].apply(lambda x : x.strip().capitalize())
dfb5['DISTRICT'] = dfb5['DISTRICT'].apply(lambda x : x.strip().capitalize())
dfb6['DISTRICT'] = dfb6['DISTRICT'].apply(lambda x : x.strip().capitalize())
dfb7['DISTRICT'] = dfb7['DISTRICT'].apply(lambda x : x.strip().capitalize())
dfb8['DISTRICT'] = dfb8['DISTRICT'].apply(lambda x : x.strip().capitalize())
```

# Egg Production Analysis

```python
data_5 = dfb5.set_index("DISTRICT", drop = False)
```

```python
# Total production egg production in 75 district
egg_production= dfb5[['DISTRICT','LAYING HEN','LAYING DUCK','HEN EGG','DUCK EGG']]

#Reshaping
egg_reshape = pd.melt(egg_production, id_vars='DISTRICT',var_name='Region',value_name='Egg Production')
#plot the data in swarmplot
plt.figure(figsize=(8,8))
sns.swarmplot(x= "Region", y= "Egg Production", data=egg_reshape)
plt.title('Swarm Plot - Egg production in 75 district')
plt.xlabel('Egg Production type')
```

# Wool Production Analysis 

```python
df_sheep= dfb7[dfb7['DISTRICT'].str.startswith(("Fw.","Mw.","W.","E.","C."))]
df_sheep
```

```python
#Total wool production in different region in Nepal
wool_production = df_sheep[['DISTRICT','SHEEPS NO.','SHEEP WOOL PRODUCED']]

#Reshaping
wool_reshape= pd.melt(wool_production, id_vars="DISTRICT",value_name="Wool Production",var_name="Sheep and Wool")
plt.figure(figsize=(20,5))

#barplot to visualize the data
sns.barplot(x="DISTRICT", y="Wool Production", data= wool_reshape, hue="Sheep and Wool")
plt.title('Bar Diagram to show wool production in different region of Nepal')
plt.xlabel('Development Region')
```

# Yak Population Analysis

```python
df_yak= dfb8[dfb8['DISTRICT'].str.startswith(("Fw.","Mw.","W.","E.","C."))]
df_yak
```

```python
import seaborn as sns
# Total yak production in different development region in Nepal.
yak_production = df_yak[['DISTRICT','YAK/NAK/CHAURI']]

# Reshaping.
yak_reshape = pd.melt(yak_production, id_vars = "DISTRICT", value_name = "Yak Production", var_name = 'Yak Type')
plt.figure(figsize=(20,5))

# Line Plot.
sns.lineplot(x ="DISTRICT", y ="Yak Production", data = yak_reshape, hue = 'Yak Type')

# Adding Title.
plt.title('Line PLot-Yak Production in Different Region of Nepal') 
plt.xlabel('Development Region')
```
